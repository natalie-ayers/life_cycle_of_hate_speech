```python
import pandas as pd
merged_df = pd.read_csv("twitter_info_annotated_racism_sexism.csv")
```

```python
# Check NA value

merged_df_na_row = merged_df[merged_df['Tweet Id'].isnull()]
merged_df_na_row
```

```python
# reference: https://github.com/JustAnotherArchivist/snscrape/issues/291

def retrive_tweets(tweet_id_lst):
    """
    Input:
        tweet_id_lst: list of tweet ids
    Output:
        Pandas DataFrame with tweet info
    """
    # Creating list to append tweet data to
    tweets_list = []
    for tweet_id in tweet_id_lst:
        # Using TwitterTweetScraper to scrape data and append tweets to list
        try:
            for i,tweet in enumerate(sntwitter.TwitterTweetScraper(tweetId=tweet_id,mode=sntwitter.TwitterTweetScraperMode.SINGLE).get_items()):
                print(tweet)
                tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.place, tweet.coordinates, tweet.user.location])
        except Exception as e:
            continue
            
    #Creating a dataframe from the tweets list above
    tweets_df = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username', 'Place', 'Coordinates', 'UserLocation'])

    #Display first 5 entries from dataframe
    tweets_df.head()
    
    return tweets_df
```

```python
prev_na_ids = merged_df_na_row['tweet_id'].tolist()
prev_na_tweet = retrive_tweets(prev_na_ids)
```
